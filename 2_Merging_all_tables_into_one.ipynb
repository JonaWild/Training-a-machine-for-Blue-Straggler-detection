```python
import numpy as np
import pandas as pd
import os

# Pfad zum Verzeichnis mit den Histogrammdateien
directory = r'C:\Users\jonaw\Projektpraktikum MPIA\BS\files'

# Liste zum Speichern der Daten aus allen Dateien
all_data = []

# Durchlaufe alle Dateien im Verzeichnis
for filename in os.listdir(directory):
    if filename.endswith(".npy"):  # Prüfe, ob die Datei eine .npy Datei ist
        filepath = os.path.join(directory, filename)
        
        # Lese die .npy Datei ein
        histogram_values = np.load(filepath)
        all_data.append(histogram_values)

# Erstelle ein DataFrame aus den gesammelten Daten
df = pd.DataFrame(all_data)

# Speichere das DataFrame in eine CSV-Datei
#df.to_csv('zusammengefuegte_histogramme.csv', index=False)

#print("Histogrammdaten erfolgreich zusammengeführt und gespeichert.")
```

```python
%%time
from sklearn.model_selection import train_test_split

Features = np.arange(100)

X = df[Features]
Y = df[100]

# first reserve 70% of the data for training, 30% for validation
X_train, X_validate, y_train, y_validate = train_test_split(X, Y, test_size=0.3, random_state=42)

# second, split the validation set in half to obtain validation and test sets.
X_validate, X_test, y_validate, y_test = train_test_split(X_validate, y_validate, test_size=0.5, random_state=42)
```

```python
%%time
from sklearn.ensemble import HistGradientBoostingClassifier
# Erstellen und Trainieren des Modells
clf = HistGradientBoostingClassifier(max_iter=900)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
%%time
import numpy as np
import matplotlib.pyplot as plt
from joblib import Parallel, delayed
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import log_loss

# Assuming X_train, X_test, y_train, and y_test are already defined and writable

# Ensure arrays are writable
X_train = np.array(X_train, copy=True)
X_test = np.array(X_test, copy=True)
y_train = np.array(y_train, copy=True)
y_test = np.array(y_test, copy=True)

def compute_log_loss(max_iter, X_train, y_train, X_test, y_test):
    params = {
        "min_samples_leaf": 5,
    }
    clf = HistGradientBoostingClassifier(max_iter=max_iter, **params)
    clf.fit(X_train, y_train)

    y_predict_train = clf.predict_proba(X_train)
    y_predict_test = clf.predict_proba(X_test)

    log_loss_values = (
        max_iter,
        log_loss(y_train, y_predict_train),
        log_loss(y_test, y_predict_test),
    )
    return log_loss_values

max_iters = np.arange(10, 200, 20).astype(int)

results = Parallel(n_jobs=-1)(
    delayed(compute_log_loss)(n_iter, X_train, y_train, X_test, y_test)
    for n_iter in max_iters
)

results = np.array(results)

fig, ax = plt.subplots(1, 1)

ax.plot(results[:, 0], results[:, 1], color="blue", label="Training Set")
ax.plot(results[:, 0], results[:, 2], color="orange", label="Test Set")

final_interval = results[:, 0] >= 100
#ax.hlines(
#    [results[final_interval, 1].mean()],
#    xmin=results[:, 0].min(),
#    xmax=results[:, 0].max(),
#    color="blue",
#    alpha=0.5,
#    linestyle="--",
#    lw=2,
#)
#ax.hlines(
#    [results[final_interval, 2].mean()],
#    xmin=results[:, 0].min(),
#    xmax=results[:, 0].max(),
#    color="orange",
#    alpha=0.5,
#    linestyle="--",
#    lw=2,
#)

ax.set_xlabel("Number of Iterations")
ax.set_ylabel("Log Loss")
ax.legend()

plt.show()

```

```python
%%time
max_iters = np.arange(10, 5000, 20).astype(int)  # Adjust the range as needed

results = Parallel(n_jobs=-1)(
    delayed(compute_log_loss)(n_iter, X_train, y_train, X_test, y_test)
    for n_iter in max_iters
)

results = np.array(results)

fig, ax = plt.subplots(1, 1)

ax.plot(results[:, 0], results[:, 1], color="blue", label="Training Set")
ax.plot(results[:, 0], results[:, 2], color="orange", label="Test Set")

final_interval = results[:, 0] >= 100

ax.set_xlabel("Number of Iterations")
ax.set_ylabel("Log Loss")
ax.legend()

plt.show()
```

```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import HistGradientBoostingClassifier
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance


result = permutation_importance(clf, X_train, y_train, n_repeats=10, random_state=42, n_jobs=-1)

# Plot feature importances
sorted_idx = result.importances_mean.argsort()
plt.figure(figsize=(10, 20))
plt.barh(range(len(sorted_idx)), result.importances_mean[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(Features)[sorted_idx])
plt.xlabel("Permutation Importance")
plt.show()
```

```python
from sklearn.inspection import PartialDependenceDisplay
top_features = np.array(Features)[sorted_idx][-5:]  # Top 5 features based on importance
fig, ax = plt.subplots(figsize=(12, 8))
PartialDependenceDisplay.from_estimator(clf, X_train, features=top_features, ax=ax, feature_names=[f"Feature {i}" for i in Features])
plt.show()
```

```python
import shap

# Create object that can calculate shap values
explainer = shap.Explainer(clf, X_train)
shap_values = explainer(X_train)

# Summary plot
shap.summary_plot(shap_values, X_train, feature_names=[f"Feature {i}" for i in Features])
```
